**Python Function**

Two types of functions. 
1. In-built
2. User defined

Same execution tasks are shown firstly by using an UDF (user Defined Fuction) and then by using in-built function.

**Calculate the length a given sequence 'TGTTCAGTCCATGACTACGTACTTTAGC' firsly by UDF and then by in-built function**

User Defined Function (UDF)

The basic steps to create a UDF:

1. Define the function by ***def*** keyword followed by name of the function user wants to give.
2. End the function definition by parenthesis ( ). Within parenthesis use parameters user wants to put [Parameters basically consist inputs].
3. Add ':' afetr the end of the line.
4. Press enter and the cursor will automatically find the indentation from where to start with the next line.
5. In the next line, add your instructions (codes) that the function would execute.
6. Finally end your function instructions with ***return*** statement. Without return statement, the function will only return the object not the desired output.

Example provided below 👇

```python
# Definie the function seq_len
def seq_len(sequence):
  counter = 0
  while sequence[counter:]:
    counter +=1
  return counter
dna_seq = 'TGTTCAGTCCATGACTACGTACTTTAGC'
print(seq_len(dna_seq))
print("The length of the given DNA sequence is: "+str(seq_len(dna_seq)))
```

```python
# Using in built function
# in-built function len()

my_dna = 'TGTTCAGTCCATGACTACGTACTTTAGC'
print(len(my_dna))
print("The length of the given DNA sequence is: "+str(len(my_dna)))
```

The data taken here, is gene expression data of several genes per sample which has been generated through transcriptomic profiling.

The corresponding article was published in Genome Biology journal and the DOI is here: https://doi.org/10.1186/gb-2013-14-10-r110 

Title: [Modeling precision treatment of breast cancer; Daemen *et. al.*](http://genomebiology.com/2013/14/10/R110)

```python
# load data
# 1000geneExpression.xslx (from .csv)
# 1000 ENSEMBL gene-IDs in rows with corresponding sample (5 Non-malignant + 6 Claudin-low) columns

import pandas as pd
gene_expression_data = pd.read_table("/content/Final_cell_lines_RNA-expression_FPKM_values_1000genes_with_NA.txt", sep = '\t', header = 0)
gene_expression_data.head(20) 
```

First, searching for any missing data. This is called data cleaning. Always look at the data and see if any data cleaning is required or not. 

Then take log normalization of the data to better plotting. Log normalization will make the data more normally distributed which helpful to understand the data pattern better.

**So what exactly I am doing here with this data ?**

1. I can there are some 'noise' in the data. Like, somewhere it is written 'NA' instead of any numerical value (index 15 claudin-low) or somewhere only interger data is given while most of the numerics are in decimal (float type). Since I have to change the enrire dataset into log scale, hence it's better to first transform the entire numeric dataset into float type from integer.

2. But again, we have a column with string values i.e. ID column. So, first ID column has to be removed to make the entire dataset numeric.

So, I have to, a) remove the 'id' column, b) convert the data into float, c) transform the dataset into log normalised.

```python
gene_expression_data['id'].head(10)
```

```python
# remove 'id' column
gene_expression_data_id_removed = gene_expression_data.drop(['id'], axis = 1)
gene_expression_data_id_removed.head(20)
```

```python
# convert numeric values into float from integer.

gene_expression_data_float = gene_expression_data_id_removed.astype(float)
gene_expression_data_float.head(28)
```

Before, log normalization, I am adding 1 with each value. cause there are many 0s in the dataset. Log of 0 would be infinity. So, adding 1 and then log of 1 would result only 0. Since it's logarithm scale, adding 1 to other non-zero values won't be a case to worry about.

```python
# log transfomation

import numpy as np
gene_expression_data_log = np.log(gene_expression_data_float+1)
gene_expression_data_log.describe() 
```
